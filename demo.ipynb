```python
# Testing `convert_weight` from `Weight_Scale`

weight_scale = Weight_Scale()

new_weight = weight_scale.convert_weight(25, 'g', 'oz')
print(f"25 grams is equal to {new_weight} ounces.")

new_weight = weight_scale.convert_weight(1.5, 'kg', 'g')
print(f"1.5 kilograms is equal to {new_weight} grams.")

try:
    weight_scale.convert_weight(1.5, 'kilos', 'g')
except ValueError as e:
    print("Error: " + str(e))
```

```python
# Testing `convert_volume` from `Volume_Scale`

volume_scale = Volume_Scale()

new_volume = volume_scale.convert_volume(25, 'ml', 'c')
print(f"25 milliliters is equal to {new_volume} cups.")

new_volume = volume_scale.convert_volume(1.5, 'l', 'ml')
print(f"1.5 liters is equal to {new_volume} milliliters.")

try:
    volume_scale.convert_volume(1.5, 'l', 'milliliters')
except ValueError as e:
    print("Error: " + str(e))
```

```python
from src.ingredient_detector import IngredientDetector

# Creating an instance of IngredientDetector
detector = IngredientDetector()
```

```python
# Testing isAllergen() method
print("1. Dairy-free diet checks:")
print("Is 'milk' allowed for dairy-free? " + str(not detector.isAllergen('milk', 'dairy-free')))
print("Is 'almond milk' allowed for dairy-free? " + str(not detector.isAllergen('almond milk', 'dairy-free')))
print("Is 'butter' allowed for dairy-free? " + str(not detector.isAllergen('butter', 'dairy-free')))
print("Is 'flour' allowed for dairy-free? " + str(not detector.isAllergen('flour', 'dairy-free')))

print("\n2. Vegan diet checks:")
print("Is 'egg' allowed for vegan? " + str(not detector.isAllergen('egg', 'vegan')))
print("Is 'honey' allowed for vegan? " + str(not detector.isAllergen('honey', 'vegan')))
print("Is 'tofu' allowed for vegan? " + str(not detector.isAllergen('tofu', 'vegan')))
print("Is 'chicken' allowed for vegan? " + str(not detector.isAllergen('chicken', 'vegan')))
print("Is 'butter' allowed for vegan? " + str(not detector.isAllergen('butter', 'vegan')))
```

```python
# Testing detectIngredients() method
ingredient_list = ["Flour", "Milk", "Eggs", "Sugar", "Butter", "Vanilla extract"]
results = detector.detectIngredients(ingredient_list)

print("3. Ingredient analysis for a baking recipe:")
for ingredient in ingredient_list:
    dairy_free = results["dairy-free"][ingredient]
    vegan = results["vegan"][ingredient]
    print(ingredient + ": Dairy-free " + dairy_free + ", Vegan " + vegan)
```

```python
# Testing generateRecipe() method
print("4. Generating recipes with random ingredients:")

# Test with 1 ingredient
print("\n1. With 1 ingredient:")
try:
    recipe = detector.generateRecipe(["Banana"])
    print("Recipe: " + recipe['title'])
    print("Ingredients used: " + str(recipe['ingredients_used']))
    print("Instructions:")
    print(recipe['instructions'])
except ValueError as e:
    print("Error: " + str(e))

# Test with 2 ingredients
print("\n2. With 2 ingredients:")
try:
    recipe = detector.generateRecipe(["Apple", "Cinnamon"])
    print("Recipe: " + recipe['title'])
    print("Ingredients used: " + str(recipe['ingredients_used']))
    print("Instructions:")
    print(recipe['instructions'])
except ValueError as e:
    print("Error: " + str(e))
```
